```python
def show_pie_plot(df,key):
    counts=df[key].value_counts()
    print(counts)
    counts_values= counts.values
    counts_labels= counts.index
    fig= plt.figure(figsize=(10,7))
    plt.pie(counts_values, labels= counts_labels)
    plt.show()
#number of matches played with different opposition
show_pie_plot(df, "Opposition")
```

```python
#number of matches played with different grounds
show_pie_plot(df, "Ground")
```

```python
#Total sixes scored in different positions
runs_at_pos= df.groupby("Pos")["6s"].sum()
print(runs_at_pos)
runs_at_pos_values= runs_at_pos.values
runs_at_pos_labels= runs_at_pos.index
fig= plt.figure(figsize=(10,7))
plt.bar(
    runs_at_pos_values,
    runs_at_pos_labels,
    color="green",
    width=4
)
plt.show()
```

```python
#Total runs scored with different countries
country_runs= df.groupby("Opposition")["Runs"].sum()
print(country_runs)
country_runs_labels= country_runs.index
country_runs_values= country_runs.values
fig= plt.figure(figsize=(15,7))
plt.bar(
    country_runs_labels,
    country_runs_values,
    color="red",
    width=0.2
)
plt.show()
```

```python
#Number of centuries scored by him in 1st and 2nd innings
centuries= df.query("Runs>=100")
print(centuries)

innings=centuries["Inns"].value_counts()
# tons= centuries["Runs"]
plt.pie(innings.values, labels=innings.index)
plt.show()
```
